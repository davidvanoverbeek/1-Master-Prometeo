# Funciones Built-in


* 1. Print

```python
print("Hola a todo el mundo")
print("Tengo", 20, " años")
```

* 2. Len

```python
print(len("Hola"))
print(len([1,2,3,4,5]))
print(len(["a", "b", "c"]))
```

* 3. Type

```python
print(type("Hola"))
print(type(True))
print(type(25))
```

* 4. Help y Dir

```python
texto = "Adios"

help(len)
help(texto.upper)

#Si ha
print(dir(texto))
```

* 5. Conversion de tipo de dato

```python
numero_en_string = "123"
print(type(numero_en_string))
numero_convertido = int(numero_en_string)
print(type(numero_convertido))

print(type(float(15.5)))
print(type(str(787))) # 787
print(type(bool(1)))
```

* 6. Funciones con numeros

```python
print(max(9,10,11,12,13,14,15))
print(min(9,10,11,12,13,14,15))
print(round(3.141516, 3))
print(abs(-7))
```

* 7. Funciones de listas

```python
numeros = [9,5,10,12,5,8,3]
print(sum(numeros))
print(sorted(numeros))
print(reversed(numeros))
```

* 8. Entrada de datos

```python
nombre = input("¿Como te llamas?")
print("El usuario se llama: ", nombre)

edad = int(input("¿Cual es tu edad actual?"))
print("EL año que viene tendras: ", edad + 1)
```

* 9. Funciones logicas

```python
print(any([0,1,0]))
print(all([1,1,1]))
print(all([1,0,1]))
```
