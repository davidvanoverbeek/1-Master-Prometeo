# Operadores Aritmeticos


### Operador de asignacion

```python
a = 10
b = 5

print(a + b)

# -SUMA
print(a + b)

# -RESTA
print(a - b)

# -MULTIPLICACION
print(a * b)

# -DIVISION
print(a / b)

# -DIVISION ENTERA
print(a // b)

# -MODULO
print(a % b)

# -POTENCIA
print(a ** b)
```

### Concatenaciones

```python
saludo = "HOla"
nombre = "Byron"
mensaje = saludo + " " + nombre
print(mensaje)

edad = 5
segundo_mensaje = nombre + " tiene " + str(edad) + " años."
print(segundo_mensaje)
```

### Operadores de comparacion o logicos

```python
x = 10
y = 20

# -Igual que
print(x == y)

# -Distinto que
print(x != y)

# -Mayor que
print(x > y)

# -Mayor igual que
print(x >= y)

# -Menor que
print(x < y)

# -Menor igual que
print(x <= y)
```

### Operadores de comparacion o logicos -> and y or

```python
nota = 7
asistencia = 0.8 #80%

aprobado = (nota > 5) and (asistencia >= 0.7)
print(aprobado)
```

### Operadores de operacion y asignacion - Combinados

```python
numero = 10
numero += 5
print(numero)

numero *= 2
print(numero)

numero /= 3
print(numero)

numero -= 2
print(numero)
```

### Conversiones de datos

```python
numero = int("5")
print(numero, type(numero))

decimal = float("3.4")
print(decimal, type(decimal))

texto = str(10)
print(texto, type(texto))

booleano = bool(0.1)
print(booleano, type(booleano))
```
