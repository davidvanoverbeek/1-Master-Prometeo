# Indexacion de strings


### Indices Positivos

```python
palabra = "Perro"

print(palabra[0])
print(palabra[3])
```

### Indices Negativos

```python
palabra = "Perro"

print(palabra[-1])
print(palabra[-5])
```

### Rango de Indexacion

```python
# string[start:stop:step]
texto = "Google Meets"

print(texto[1:4])
print(texto[:4])
print(texto[7:])
print(texto[0:6:2])
print(texto[::2])
print(texto[::-1])
```

### Metodo de string

```python
pelicula = "Stars Wars"
print(pelicula.lower())
print(pelicula.upper())

pelicula = "el caballero oscuro"
print(pelicula.title())
print(pelicula.capitalize())

pelicula = " Cantando bajo la lluvia    "
print(pelicula.strip())
print(pelicula.replace("lluvia", "luna"))
print(pelicula.find("bajo"))
print(pelicula.find("perro"))
print(pelicula.index("bajo"))
```

### Lista

```python
lista = ["Perro", "Gato", "Loro", "Ciervo"]
print(lista)
print(" | ".join(lista))
print(" , ".join(lista))
print("+".join(lista))
```
