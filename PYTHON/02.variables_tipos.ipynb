# Variables

```python
# Declarar y asignar el valor de una variable con el operador de asignacion = 
variable = 5
print(variable)

# Reasignar un valor a una variable que ya existe
variable = 4
print(variable)
```

### Tipo de Dato

##### Numero

```python
# int
age = 18
print(type(age))

#float
price = 10.25
print(type(price))

#complex
complejo = 2 + 4j
print(type(complejo))
```

##### Strings


```python
nombre = "Antonio"
nombre = 'Antonio'
frase = 'Como bien dijo Rajoy: "¡Viva el vino!"'
print(frase)

varias_lineas = '''
Este es un string
que se muestra
en varias
lineas
'''
print(varias_lineas)

caracter_de_escape = "Buenas tardes a \n'todos\'"
print(caracter_de_escape)
print(type(caracter_de_escape))

print(f"La variable nombre tiene como valor: {nombre}")
```

##### Booleanos

```python
aprobado = True
matriculado = False
print(type(aprobado))
```

##### None

```python
usuario = None
print(type(usuario))
```

### Ejemplo de funcion built-in

##### Len

```python
ejemplo_string = "Esto es una frase"
longitud_string = len(ejemplo_string)
print(longitud_string)

print(f'La longitud de la frase: "{ejemplo_string}" es: {len(ejemplo_string)}')
```
